```python
from pathlib import Path

import pandas as pd


class DatasetAnalyzer:

    def __init__(self, parquet_path: str):

        self.parquet_path = Path(parquet_path)

        self.df = pd.read_parquet(self.parquet_path)

    def basic_statistics(self):

        stats = {

            "total_reviews": len(self.df),

            "total_products":
                self.df["product_id"].nunique(),

            "total_customers":
                self.df["customer_id"].nunique(),

            "average_rating":
                round(self.df["rating"].mean(), 2),

            "average_review_length":
                round(self.df["word_count"].mean(), 2),

            "average_helpful_votes":
                round(self.df["helpful_total"].mean(), 2),
        }

        return stats
    
    def missing_values(self):

        missing = (
            self.df
            .isnull()
            .sum()
            .sort_values(ascending=False)
        )

        return missing.to_dict()


    def sentiment_distribution(self):

        return (
            self.df["sentiment_label"]
            .value_counts(normalize=True)
            .mul(100)
            .round(2)
            .to_dict()
        )

    def rating_distribution(self):

        return (
            self.df["rating"]
            .value_counts()
            .sort_index()
            .to_dict()
        )

    def category_distribution(self):

        return (
            self.df["category"]
            .fillna("Unknown")
            .value_counts()
            .head(20)
            .to_dict()
        )

    # def brand_distribution(self):

    #     return (
    #         self.df["brand"]
    #         .fillna("Unknown")
    #         .value_counts()
    #         .head(20)
    #         .to_dict()
    #     )
```

```python
from pprint import pprint

analyzer = DatasetAnalyzer(
    "../../datasets/processed/master_reviews.parquet"
    
)

print("\n===== BASIC STATS =====")
pprint(analyzer.basic_statistics())

print("\n===== SENTIMENT =====")
pprint(analyzer.sentiment_distribution())

print("\n===== RATINGS =====")
pprint(analyzer.rating_distribution())

print("\n===== CATEGORIES =====")
pprint(analyzer.category_distribution())

print("\n===== MISSING =====")
pprint(analyzer.missing_values())
```
